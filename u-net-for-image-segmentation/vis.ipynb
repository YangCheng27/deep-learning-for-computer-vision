```python
import unet
import matplotlib.pyplot as plt
import kagglehub

path = kagglehub.dataset_download("maedemaftouni/covid19-ct-scan-lesion-segmentation-dataset")
loaders = unet.get_dataloaders(path)
```

```python
unet.data_vis(loaders['evaluation'])
```

```python
model = unet.SimpleUNet()
```

```python
train_losses, val_losses, epoch_times, train_metrics_history, val_metrics_history = unet.train_model(
    model, loaders['training'], loaders['evaluation'], num_epochs=30, device=unet.get_device()
)
```

```python
unet.plot_result(train_losses, val_losses, train_metrics_history, val_metrics_history)
```
